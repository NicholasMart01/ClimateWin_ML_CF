# Importing Libraries and Data 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
path = r'C:\Users\Administrator\Desktop\Projects (Carrer foundry)\Data sets'
df = pd.read_csv(os.path.join(path, 'Dataset-weather-prediction-dataset-processed.csv'))
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
df.rename(columns={'index':'id'}, inplace=True)
df.head()
```

# Scaling Data 

```python
df2 = df.drop(columns=['id','DATE','MONTH'])
df2.head()
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df2)
                        )
df_scaled.head()
```

```python
df_subset = df[['id','DATE','MONTH']]
df_scaled2 = pd.concat([df_subset, df_scaled], axis=1)
```

```python
df_scaled2.columns = df.columns
df_scaled2.head()
```

```python
df_scaled2.to_csv(os.path.join(path, 'DataSet_scaled.csv'))
```
