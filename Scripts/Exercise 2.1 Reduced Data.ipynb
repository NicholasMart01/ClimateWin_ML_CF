# Importing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib.pyplot import figure

from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
path = r'C:\Users\Administrator\Desktop\Projects (Carrer foundry)\Data sets'
```

```python
reduced = pd.read_csv(os.path.join(path, '2019_PCA.csv'))
```

# Dendrograms

```python
reduced_single = linkage(reduced,method="single")
plt.figure(figsize=(18,6))
dendrogram(reduced_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method",fontsize=18)
plt.show()
```

```python
reduced_complete = linkage(reduced,method="complete")
plt.figure(figsize=(18,6))
dendrogram(reduced_complete, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Complete Method",fontsize=18)
plt.show()
```

```python
reduced_average = linkage(reduced,method="average")
plt.figure(figsize=(18,6))
dendrogram(reduced_average, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Average Method",fontsize=18)
plt.show()
```

```python
reduced_ward = linkage(reduced,method="ward")
plt.figure(figsize=(18,6))
dendrogram(reduced_ward, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Ward Method",fontsize=18)
plt.show()
```
